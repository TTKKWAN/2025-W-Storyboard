```python
import os
import cv2
import torch
import numpy as np
from tqdm import tqdm
from PIL import Image
from ultralytics import YOLO  # YOLOv8-Segmentation

# ====== 1️⃣ 경로 설정 ======
input_dir = "/content/drive/MyDrive/머신러닝/2025-2 d2d/웹툰 자동화 프로젝트/Training_1024x1024"
output_dir = "/content/drive/MyDrive/머신러닝/2025-2 d2d/웹툰 자동화 프로젝트/BubbleRemoved_Edges"
os.makedirs(output_dir, exist_ok=True)

# ====== 2️⃣ 이미지 리스트 ======
image_files = [f for f in os.listdir(input_dir)
               if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

print(f"총 {len(image_files)}개의 이미지 변환 시작...\n")

# ====== 3️⃣ YOLOv8 말풍선 세그멘테이션 모델 로드 ======
# huggingface: kitsumed/yolov8m_seg-speech-bubble
model = YOLO("/content/drive/MyDrive/머신러닝/2025-2 d2d/웹툰 자동화 프로젝트/model.pt")


stats = {'success': 0, 'bubbles_removed': 0, 'no_bubbles': 0, 'failed': 0}

# ====== 4️⃣ 루프 ======
for file_name in tqdm(image_files, desc="Processing"):
    file_path = os.path.join(input_dir, file_name)

    try:
        img = cv2.imread(file_path)
        if img is None:
            print(f"\n⚠️ {file_name} 로드 실패")
            stats['failed'] += 1
            continue

        h, w, _ = img.shape

        # ----- (1) YOLOv8 예측 -----
        results = model.predict(source=img, conf=0.4, verbose=False)
        masks = results[0].masks

        # ----- (2) 마스크 생성 -----
        if masks is not None:
            mask_data = masks.data.cpu().numpy()  # [N, H, W]
            mask = (np.sum(mask_data, axis=0) > 0).astype(np.uint8) * 255
            mask = cv2.resize(mask, (w, h))

            # ----- (3) 마스크 후처리 -----
            kernel = np.ones((5, 5), np.uint8)
            mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel, iterations=2)

            # ----- (4) 말풍선 제거 (inpainting) -----
            result = cv2.inpaint(img, mask, 3, cv2.INPAINT_TELEA)
            stats['bubbles_removed'] += 1
        else:
            result = img.copy()
            stats['no_bubbles'] += 1

        # ----- (5) Canny Edge -----
        gray = cv2.cvtColor(result, cv2.COLOR_BGR2GRAY)
        gray = cv2.GaussianBlur(gray, (5, 5), 1.0)
        median = np.median(gray)
        lower = int(max(0, 0.7 * median))
        upper = int(min(255, 1.3 * median))
        edges = cv2.Canny(gray, lower, upper)
        edges_inv = cv2.bitwise_not(edges)

        # ----- (6) 저장 -----
        save_path = os.path.join(output_dir, f"edge_{file_name}")
        Image.fromarray(edges_inv).save(save_path)
        stats['success'] += 1

    except Exception as e:
        print(f"\n❌ {file_name} 오류: {e}")
        stats['failed'] += 1

# ====== 5️⃣ 결과 요약 ======
print("\n" + "="*50)
print("🎨 처리 완료!")
print("="*50)
print(f"✅ 성공: {stats['success']}개")
print(f"💬 말풍선 제거: {stats['bubbles_removed']}개")
print(f"📄 말풍선 없음: {stats['no_bubbles']}개")
print(f"❌ 실패: {stats['failed']}개")
print(f"\n📂 출력 폴더: {output_dir}")
```

```python
import torch
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer
import os

# 모델 로드
MODEL = "gyupro/Koalpaca-Translation-KR2EN"

# 임시 오프로드 폴더 생성
offload_folder = "/tmp/offload"
os.makedirs(offload_folder, exist_ok=True)

model = AutoModelForCausalLM.from_pretrained(
    MODEL,
    torch_dtype=torch.float16,
    device_map="auto",
    offload_folder=offload_folder # offload_folder 추가
)
tokenizer = AutoTokenizer.from_pretrained(MODEL)

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)
```

```python
# 입력 디렉토리 및 출력 디렉토리 설정
input_dir = '/content/drive/MyDrive/머신러닝/2025-2 d2d/웹툰 자동화 프로젝트/text'
output_dir = '/content/drive/MyDrive/머신러닝/2025-2 d2d/웹툰 자동화 프로젝트/translated_text'

# 출력 디렉토리 생성 (없으면)
os.makedirs(output_dir, exist_ok=True)

# 입력 디렉토리의 모든 파일 목록 가져오기
text_files = [f for f in os.listdir(input_dir) if os.path.isfile(os.path.join(input_dir, f)) and f.endswith('.txt')]

# 각 파일 번역
for file_name in text_files:
    input_path = os.path.join(input_dir, file_name)
    output_path = os.path.join(output_dir, f"translated_{file_name}") # 번역된 파일 이름 설정

    try:
        # 입력 파일 읽기
        with open(input_path, "r", encoding="utf-8") as f:
            lines = [line.strip() for line in f.readlines() if line.strip()]

        translated_lines = []

        # 한 줄씩 번역
        print(f"Translating file: {file_name}")
        for i, text in enumerate(lines):
            print(f"  Translating line {i+1}/{len(lines)}...")
            ans = pipe(
                f"### source: {text}\n\n### target:",
                do_sample=False,
                max_new_tokens=512,
                temperature=0.7,
                top_p=0.9,
                return_full_text=False,
                eos_token_id=2,
            )
            msg = ans[0]["generated_text"]

            # '###' 이후의 불필요한 텍스트 제거
            if "###" in msg:
                msg = msg.split("###")[0]

            translated_lines.append(msg.strip())

        # 결과 저장
        with open(output_path, "w", encoding="utf-8") as f:
            for src, tgt in zip(lines, translated_lines):
                f.write(f"{src}\n→ {tgt}\n\n")

        print(f"✅ Translation complete for {file_name}. Result saved to {output_path}")

    except Exception as e:
        print(f"Error processing file {file_name}: {e}")


print("✅ All files processed.")
```
